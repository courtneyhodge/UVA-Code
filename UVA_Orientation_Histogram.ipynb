```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/MSDS-Orientation-Computer-Survey.csv")
df
```

```python
val_counts = df['CPU Number of Cores (int)'].value_counts()
val_counts
```

```python
plt.figure(figsize = (10, 6))
val_counts.plot(kind = 'bar', color = 'olive', edgecolor = 'white')

plt.title('CPU Number of Cores', fontsize = 23)
plt.xlabel('CPU Number of Cores (int)', fontsize = 15)
plt.ylabel('Count', fontsize = 15)


plt.show()
```
